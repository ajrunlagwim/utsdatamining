## 1.Lakukan Preprocessing pada dataset dengan menghilangkan variabel yang tidak digunakan dan mengubah dokumen dari file XLSX ke file CSV

```python
import pandas as pd

data = pd.read_excel('/content/drive/MyDrive/dataset data mining uts/DatasetCuaca_convert.xlsx')
```

```python
data.head
```

```python
selected_columns = ['Bulan dan Tahun', 'Suhu Udara',  'Kelembapan Udara',  'Curah Hujan']
data = data[selected_columns]
```

```python
data.head
```

```python
data.to_csv('convert dataset.csv', index=False)
```

## 2.       Mencari nilai Mean, Median dan Modus dari Suhu Udara


```python
import pandas as pd

# data = pd.read_csv('convert dataset.csv')
suhu_udara = data['Suhu Udara']

# Mean
mean_suhu_udara = suhu_udara.mean()

# Median
median_suhu_udara = suhu_udara.median()

# Modus
modus_suhu_udara = suhu_udara.mode().values[0]

print("Mean Suhu Udara:", mean_suhu_udara)
print("Median Suhu Udara:", median_suhu_udara)
print("Modus Suhu Udara:", modus_suhu_udara)
```

## 3.       Mencari nilai Mean, Median dan Modus dari Curah Hujan

```python
import pandas as pd

# data = pd.read_csv('nama_file.csv')
curah_hujan = data['Curah Hujan']

# Mean
mean_curah_hujan = curah_hujan.mean()

# Median
median_curah_hujan = curah_hujan.median()

# Modus
modus_curah_hujan = curah_hujan.mode().values[0]

print("Mean Curah Hujan:", mean_curah_hujan)
print("Median Curah Hujan:", median_curah_hujan)
print("Modus Curah Hujan:", modus_curah_hujan)
```

## 4.       Eksplorasi Data Analysis (EDA) untuk Data Kelembaban Udara dan Curah Hujan. Gunakan Boxplot dan Scatterplot untuk memvisualisasikan Data tersebut.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# data = pd.read_csv('nama_file.csv')

# Boxplot Kelembaban Udara
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, y='Kelembapan Udara')
plt.title('Boxplot Kelembaban Udara')
plt.ylabel('Kelembapan Udara')
plt.show()

# Scatterplot Kelembaban Udara vs Curah Hujan
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Kelembapan Udara', y='Curah Hujan')
plt.title('Scatterplot Kelembaban Udara vs Curah Hujan')
plt.xlabel('Kelembapan Udara')
plt.ylabel('Curah Hujan')
plt.show()
```

## 5.       Visualisasikan Standar Deviasi pada data Curah Hujan dan Suhu menggunakan Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

# data = pd.read_csv('nama_file.csv')

# Standar deviasi Curah Hujan
std_curah_hujan = data['Curah Hujan'].std()

# Standar deviasi Suhu Udara
std_suhu = data['Suhu Udara'].std()

# Histogram Curah Hujan
plt.figure(figsize=(8, 6))
plt.hist(data['Curah Hujan'], bins=10, alpha=0.5)
plt.axvline(x=std_curah_hujan, color='r', linestyle='--', linewidth=2, label='Standar Deviasi')
plt.title('Histogram Curah Hujan')
plt.xlabel('Curah Hujan')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()

# Histogram Suhu Udara
plt.figure(figsize=(8, 6))
plt.hist(data['Suhu Udara'], bins=10, alpha=0.5)
plt.axvline(x=std_suhu, color='r', linestyle='--', linewidth=2, label='Standar Deviasi')
plt.title('Histogram Suhu Udara')
plt.xlabel('Suhu Udara')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

## 6.       Tampilkan EDA dalam bentuk Report Non Graphical untuk untuk mengidentifikasi Pola dan menemukan Anomali Data

```python
import pandas as pd

# data = pd.read_csv('nama_file.csv')

# Statistik Deskriptif
num_entries = len(data)
time_range = f"{data['Bulan dan Tahun'].min()} - {data['Bulan dan Tahun'].max()}"

# Suhu Udara
suhu_mean = data['Suhu Udara'].mean()
suhu_std = data['Suhu Udara'].std()
suhu_range = f"{data['Suhu Udara'].min()} - {data['Suhu Udara'].max()}"

# Curah Hujan
curah_hujan_mean = data['Curah Hujan'].mean()
curah_hujan_std = data['Curah Hujan'].std()
curah_hujan_range = f"{data['Curah Hujan'].min()} - {data['Curah Hujan'].max()}"

# Identifikasi Anomali Data
max_curah_hujan = data['Curah Hujan'].max()
min_curah_hujan = data['Curah Hujan'].min()
anomali_bulan = data[(data['Curah Hujan'] == max_curah_hujan) | (data['Curah Hujan'] == min_curah_hujan)]

# Laporan EDA
report = f"EDA Report:\n\n"
report += f"Jumlah Data: {num_entries} entri\n"
report += f"Rentang Waktu: {time_range}\n\n"
report += f"Statistik Deskriptif Suhu Udara:\n"
report += f"  - Rata-rata: {suhu_mean:.2f}\n"
report += f"  - Standar Deviasi: {suhu_std:.2f}\n"
report += f"  - Rentang: {suhu_range}\n\n"
report += f"Statistik Deskriptif Curah Hujan:\n"
report += f"  - Rata-rata: {curah_hujan_mean:.2f}\n"
report += f"  - Standar Deviasi: {curah_hujan_std:.2f}\n"
report += f"  - Rentang: {curah_hujan_range}\n\n"
report += f"Identifikasi Anomali Data Curah Hujan:\n"
report += f"{anomali_bulan.to_string(index=False)}\n"

print(report)
```
